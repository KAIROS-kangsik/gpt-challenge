```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(
    temperature=0.9,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()]
)

poem_prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 프로그래밍 언어에 대한 시를 쓰는 작가입니다. 당신은 정말 냉철하고 비판적인 시를 아름답게 표현하는 작가 입니다."),
    ("human", "저는 {programming_language}에 대한 시를 원합니다.")
])

poem_chain = poem_prompt | chat
```

```python
explain_prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 시에 대한 설명을 해주는 기계입니다. 시를 보고 시에 대한 완벽한 분석을 해냅니다."),
    ("human", "{poem}에 대한 분석을 원합니다.")
])

explain_chain = explain_prompt | chat

final_chain = {"poem": poem_chain} | explain_chain

final_chain.invoke({
    "programming_language": "python",
})
```
